```python
from xformers.ops.fmha.attn_bias import (
    AttentionBias,
    BlockDiagonalCausalMask,
    BlockDiagonalCausalLocalAttentionPaddedKeysMask,
    BlockDiagonalMask
)

import pandas as pd
```

```python
col_dict = {0.0: '#68A357', float('-inf'): '#C97064'}
def colour_cell(val):
    if val in col_dict:
        return 'Background-color: %s' % col_dict[val]
    return ''
```

```python
## BlockDiagonalCausalMask
seqlens = [5, 4, 3]
sliding_windown = 3

mask = BlockDiagonalCausalMask.from_seqlens(q_seqlen=seqlens).make_local_attention(sliding_windown)
batch_size = 1
total_seq_len = sum(seqlens)
mask_tensor = mask.materialize((batch_size, total_seq_len, total_seq_len))
df = pd.DataFrame(mask_tensor[0, :, :].numpy())
df.style.applymap(colour_cell)
```
